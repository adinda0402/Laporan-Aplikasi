# **Algoritma & Struktur Data**
**Queue Prioritas (Sistem Pemesanan Tiket)**

* Adinda  (233210214)
* Mutia Anggraini (22310079)
* Salmaa Candraningtyas Putri Bimantoro (233210691)





```python
import heapq

class TicketSystem:
    def __init__(self):
        self.queue = []  # Priority Queue
        self.ticket_id = 1  # Ticket ID generator
        self.available_tickets = 10  # Total tickets tersedia

    def book_ticket(self, name, priority):
        if self.available_tickets > 0:
            # Priority: VIP (1) lebih tinggi dari Regular (2)
            heapq.heappush(self.queue, (priority, self.ticket_id, name))
            self.ticket_id += 1
            self.available_tickets -= 1
            print(f"Tiket berhasil dibooking untuk {name}!")
        else:
            print("Maaf, tiket tidak tersedia.")

    def process_tickets(self):
        if self.queue:
            priority, ticket_id, name = heapq.heappop(self.queue)
            print(f"Memproses tiket {ticket_id} untuk {name}. (Prioritas: {'VIP' if priority == 1 else 'Regular'})")
        else:
            print("Tidak ada ticket untuk di proses.")

    def display_queue(self):
        if self.queue:
            print("Antrian tiket saat ini:")
            for priority, ticket_id, name in sorted(self.queue):
                print(f"Tiket ID: {ticket_id}, Nama: {name}, Prioritas: {'VIP' if priority == 1 else 'Regular'}")
        else:
            print("Tidak ada tiket di antrian.")

    def display_status(self):
        print(f"Tiket tersedia: {self.available_tickets}")
        print(f"Tiket dalam Antrian: {len(self.queue)}")

# Main Program
system = TicketSystem()

while True:
    print("\n1. Booking Tiket")
    print("2. Tiket Terproses")
    print("3. Tampilkan Antrian")
    print("4. Tampilkan Status")
    print("5. Exit")
    choice = input("Masukkan Pilihan: ")

    if choice == "1":
        name = input("Nama Anda: ")
        print("Pilihan Prioritas:")
        print("1. VIP")
        print("2. Regular")
        priority = int(input("Masukkan Prioritas (1/2): "))
        system.book_ticket(name, priority)
    elif choice == "2":
        system.process_tickets()
    elif choice == "3":
        system.display_queue()
    elif choice == "4":
        system.display_status()
    elif choice == "5":
        print("Keluar dari sistem. Thank you!")
        break
    else:
        print("Pilihan tidak valid, silahkan coba lagi.")
```
